```python
# Now that we have gotten familiar with pandas, let's think about how to aggregate data into something for visualization

bellevue_df['disease'].value_counts()
```

```python
# 75 is a lot of potential options, so let's just look at the top 10

bellevue_df['disease'].value_counts()[:10]
```

```python
# Pandas has some built in visualization functions, so we can use those to get a quick exploratory glance at our data.

bellevue_df['disease'].value_counts()[:10].plot(kind='bar', title='Bellevue Almshouse:\nMost Frequent "Diseases"')

```

It's not pretty, but gives us a rough sketch of our distributions.

```python
# Lets try something more complicated ... let's get the top professions to filter our data
# Then we can break that out by gender and format it for visualization.
```

```python
bellevue_df['profession'].value_counts().head(15).reset_index()
```

```python
# Next lets create a filter for those professions by creating a list of the job titles we want.
professions = bellevue_df['profession'].value_counts().head(15).reset_index()['index'].tolist()
print(professions)

# Create a new dataframe with only rows that correspond to our target professions.
top_15 = bellevue_df.profession.isin(professions)
top_15
```

```python
top_professions = bellevue_df[top_15]
top_professions
```

```python
# Now we can group our data by profession and by gender and get counts.

top_professions.groupby(["profession", "gender"])['last_name'].count()
```

```python
# We can also filter down the data by a particular variable

profession = top_professions.groupby(["profession", "gender"])['last_name'].count().reset_index()
```

```python
profession
```

```python
# Finally, we can shift our data from a long format to a wide one for visualizing using the "pivot" function.

profession_wide = profession.pivot(index="profession", columns="gender", values="last_name").reset_index()
profession_wide
```

```python
# And we can export the data to a CSV file

profession_wide.to_csv("profession_by_gender.csv", encoding='utf-8', index=False)
```
